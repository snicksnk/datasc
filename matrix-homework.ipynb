## Уровень 1

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
]
A = np.array(A)
x = A[:,0]
y = A[:,1]
print(x, y)
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])




```

### Уменьшаем в 3 раза

```python
ratio = 1/3
ResizeMatrix = np.array([[ratio, 0, 0], [0, ratio, 0], [0, 0, 1]])
ResizedMatrix = A @ ResizeMatrix

x = ResizedMatrix[:,0]
y = ResizedMatrix[:,1]
plt.plot(x, y)
plt.ylim([-100, 100])
plt.xlim([-100, 100])
```

### Повернуть на 130 градусов

```python
import math

fi = np.deg2rad(130)

RotateMatrix = np.array([[ math.cos(fi), math.sin(fi), 0], 
                        [-math.sin(fi), math.cos(fi), 0],
                        [0,             0,            1]])
RotatedMatrix = A @ RotateMatrix
x = RotatedMatrix[:,0]
y = RotatedMatrix[:,1]
plt.plot(x, y)
#plt.ylim([-200, 100])
#plt.xlim([-100, 100])
```

### Отразить объект относительно прямой y=x

```python
fi = np.deg2rad(180)
import math
RotateMatrix = np.array([[ math.cos(fi), math.sin(fi), 0], 
                        [-math.sin(fi), math.cos(fi), 0],
                        [0,             0,            1]])
RotatedMatrix = A @ RotateMatrix
x = RotatedMatrix[:,0]
y = RotatedMatrix[:,1]
plt.plot(x, y)
```
